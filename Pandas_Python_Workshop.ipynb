## 🐼 Core Capabilities of Pandas

Pandas is built on top of the NumPy library and offers high-performance, easy-to-use data structures and data analysis tools.

### 1\. Data Loading and Saving

Pandas allows you to easily read and write data from various formats, making it highly versatile.

*   **Loading:** It can read data from **CSV** files (the most common format), Excel sheets, SQL databases, JSON, and even plain text files. The primary function for this is `pd.read_csv()`.
    
    *   _Example:_ `df = pd.read_csv('data.csv')`
        
*   **Saving:** You can export your cleaned or processed DataFrame back into these formats using methods like `df.to_csv()`.
    

### 2\. Data Inspection and Exploration

Once data is loaded, you need tools to understand its structure and content.

*   **Viewing Data:**
    
    *   `df.head()` and `df.tail()`: Show the first and last $N$ rows, respectively, to get a quick glimpse of the data.
        
    *   `df.sample()`: Returns a random sample of rows.
        
*   **Summary Statistics:**
    
    *   `df.info()`: Provides a concise summary of the DataFrame, including the number of entries, the data type of each column (Dtype), and the count of non-null values.
        
    *   `df.describe()`: Generates descriptive statistics (count, mean, standard deviation, min, max, quartiles) of the numerical columns.
        
*   **Unique Values:**
    
    *   `df['column_name'].value_counts()`: Lists all unique values in a column and how many times each value appears, crucial for understanding categorical features.
        

### 3\. Data Cleaning and Preprocessing

This is often the most time-consuming part of machine learning, and Pandas simplifies many common tasks.

*   **Handling Missing Data (NaNs):**
    
    *   `df.isnull()`: Generates a boolean mask indicating where data is missing.
        
    *   `df.dropna()`: Removes rows or columns that contain missing values.
        
    *   `df.fillna()`: Fills missing values with a specified value (e.g., the mean, median, or a constant).
        
*   **Data Type Conversion:**
    
    *   `df['column'] = df['column'].astype(new_type)`: Used to change a column's data type, for example, converting a string column to a numeric one or an integer to a category.
        
*   **Removing Duplicates:**
    
    *   `df.drop_duplicates()`: Identifies and removes duplicate rows.
        

### 4\. Data Selection, Indexing, and Subsetting

Pandas provides intuitive ways to access specific parts of the data.

*   **Selecting Columns:** Accessing a column is similar to accessing a dictionary key: `df['column_name']`.
    
*   **Filtering Rows:** Use boolean masks for powerful conditional selection: `df[df['age'] > 30]`.
    
*   **Location-based Indexing:**
    
    *   `df.loc[]`: Primarily used for **label-based** indexing (using row/column names).
        
    *   `df.iloc[]`: Primarily used for **integer position-based** indexing (using numerical indices).
        

### 5\. Grouping and Aggregation

This allows for applying functions to subsets of the data, similar to the "Group By" clause in SQL.

*   **The `groupby()` method:** Splits the DataFrame into groups based on some criterion.
    
    *   Example: To find the average salary per department:
        
        df.groupby('Department')\['Salary'\].mean()
        
    *   Common aggregation functions include `mean()`, `sum()`, `count()`, `min()`, and `max()`.

```python
import pandas as pd
from IPython.display import display

# -------- Load DataFrame --------
df = pd.read_csv("employees.csv")
display(df)

```

```python
# -------- Data Overview --------
df.info()
display(df.head())
display(df.describe())
```

```python
# -------- Missing Data Handling --------
print(df.isnull().sum())

df_dropna = df.dropna()
df_fillna_zero = df.fillna(0)
df_fillna_mean = df.fillna(df.mean(numeric_only=True))

display(df_dropna)
display(df_fillna_zero)
display(df_fillna_mean)
```

```python
# -------- Data Cleaning Techniques --------

# Remove missing rows
df.dropna(inplace=True)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing numeric values with mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill missing string values with a placeholder
df.fillna("Unknown", inplace=True)

# Convert data types
df["Age"] = df["Age"].astype(int)

# Strip extra spaces from strings
df["First Name"] = df["First Name"].str.strip()

# Convert text to uniform case
df["Department"] = df["Department"].str.upper()

# Replace incorrect or inconsistent values
df.replace("IT DEPT", "IT", inplace=True)

# Remove unwanted characters (example)
df["First Name"] = df["First Name"].str.replace("[^a-zA-Z ]", "", regex=True)

# Reset index after cleaning
df.reset_index(drop=True, inplace=True)

display(df)
```

```python
df.to_csv("cleaned_employees.csv", index=False)
```

```python

# -------- Basic Statistics --------
print(df["Monthly_Salary"].mean())
print(df["Monthly_Salary"].max())
print(df["Monthly_Salary"].min())
print(df["Monthly_Salary"].std())
```
