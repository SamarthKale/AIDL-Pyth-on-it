## **Introduction to Matplotlib**

Matplotlib is a widely used **Python library for data visualization**. It allows us to represent numerical data in the form of **graphs and charts**, making complex data easier to understand and analyze.

In real-world data analysis and machine learning, data is often available as large tables of numbers. Reading such raw numerical data can be confusing and time-consuming. Matplotlib helps by converting these numbers into visual forms such as **line graphs, bar charts, scatter plots, histograms, and pie charts**, which clearly show patterns, trends, comparisons, and relationships.

Matplotlib integrates smoothly with **Pandas and NumPy**, making it an essential tool in the data science workflow. Typically, data is first loaded and cleaned using Pandas, numerical operations are performed using NumPy, and then Matplotlib is used to visualize the results.

In this workshop, Matplotlib is used to:

- Visualize employee data such as age and salary  
- Display relationships between variables  
- Show the **best-fit line** in Linear Regression  
- Help students visually understand how a machine learning model learns patterns  

Matplotlib is beginner-friendly, flexible, and works across different environments such as **VS Code, Google Colab, and Jupyter Notebook**. It plays a crucial role in making data-driven decisions more intuitive and understandable.

In simple words, **Matplotlib helps us see the data instead of just reading numbers**.

## **Common Matplotlib Commands (Quick Reference Table)**

| Purpose | Command |
|------|-------|
| Import Matplotlib | `import matplotlib.pyplot as plt` |
| Create a new figure | `plt.figure()` |
| Line graph | `plt.plot(x, y)` |
| Scatter plot | `plt.scatter(x, y)` |
| Bar chart | `plt.bar(x, y)` |
| Horizontal bar chart | `plt.barh(x, y)` |
| Histogram | `plt.hist(data)` |
| Pie chart | `plt.pie(data)` |
| Box plot | `plt.boxplot(data)` |
| Set title | `plt.title("Title")` |
| Set x-axis label | `plt.xlabel("X label")` |
| Set y-axis label | `plt.ylabel("Y label")` |
| Rotate x-axis labels | `plt.xticks(rotation=45)` |
| Display grid | `plt.grid(True)` |
| Show the plot | `plt.show()` |
| Plot markers | `plt.plot(x, y, marker='o')` |
| Change color | `plt.plot(x, y, color='red')` |
| Add legend | `plt.legend()` |

## **Typical Visualization Workflow**

1. Load data using Pandas  
2. Select columns for visualization  
3. Choose an appropriate plot type  
4. Add labels and title  
5. Display the graph  

---



```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("employees.csv")
```

```python

# ---------- 1. Line Graph: Salary vs Age ----------
plt.figure()
plt.plot(df["Age"], df["Salary"], marker="o")
plt.title("Salary vs Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# ---------- 2. Scatter Plot: Salary vs Age ----------
plt.figure()
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# ---------- 3. Bar Graph: Salary Comparison ----------
plt.figure()
plt.bar(df["First Name"], df["Salary"])
plt.title("Salary by Employee")
plt.xlabel("First Name")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------- 4. Horizontal Bar Graph ----------
plt.figure()
plt.barh(df["First Name"], df["Salary"])
plt.title("Salary Comparison (Horizontal)")
plt.xlabel("Salary")
plt.ylabel("First Name")
plt.show()
```

```python
# ---------- 5. Education Level vs Average Salary ----------
plt.figure()
df.groupby("Education")["Salary"].mean().plot(kind="bar", color="teal")
plt.title("Average Salary by Education Level")
plt.xlabel("Education")
plt.ylabel("Average Salary")
plt.xticks(rotation=0)
plt.show()
```

```python
# ---------- 6. Gender-wise Salary Distribution ----------
plt.figure()
df.groupby("Gender")["Salary"].mean().plot(kind="bar", color=["pink","skyblue"])
plt.title("Average Salary by Gender")
plt.xlabel("Gender")
plt.ylabel("Salary")
plt.show()
```

```python
# ---------- 7. Pie Chart: Department Distribution ----------
plt.figure()
df["Department"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Department Distribution")
plt.ylabel("")
plt.show()
```

```python
# ---------- 8. Pie Chart: City Distribution ----------
plt.figure()
df["City"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("City Distribution")
plt.ylabel("")
plt.show()
```

```python
# ---------- 9. Histogram: Age Distribution ----------
plt.figure()
plt.hist(df["Age"], bins=8, edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# ---------- 10. Histogram: Salary Distribution ----------
plt.figure()
plt.hist(df["Salary"], bins=10, edgecolor="black")
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
# ---------- 11. Boxplot: Salary ----------
plt.figure()
plt.boxplot(df["Salary"])
plt.title("Salary Spread")
plt.ylabel("Salary")
plt.show()



```

```python
# ---------- 12. Boxplot: Age ----------
plt.figure()
plt.boxplot(df["Age"])
plt.title("Age Spread")
plt.ylabel("Age")
plt.show()
```

```python
# ---------- 13. Bonus % vs Salary Scatter ----------
plt.figure()
plt.scatter(df["Bonus %"], df["Salary"], color="purple")
plt.title("Bonus % vs Salary")
plt.xlabel("Bonus %")
plt.ylabel("Salary")
plt.show()
```

```python
# ---------- 14. Senior Management Count ----------
plt.figure()
df["Senior Management"].value_counts().plot(kind="bar", color="orange")
plt.title("Senior Management Count")
plt.xlabel("Senior Management")
plt.ylabel("Count")
plt.show()
```

```python
# ---------- 15. Joining Year vs Salary Trend ----------
plt.figure()
df.groupby("JoiningYear")["Salary"].mean().plot(marker="o")
plt.title("Average Salary by Joining Year")
plt.xlabel("Joining Year")
plt.ylabel("Average Salary")
plt.show()


```

```python
# ---------- 16. Education vs Age ----------
plt.figure()
df.groupby("Education")["Age"].mean().plot(kind="bar", color="green")
plt.title("Average Age by Education")
plt.xlabel("Education")
plt.ylabel("Average Age")
plt.show()
```

```python
# ---------- 17. EverBenched vs Salary ----------
plt.figure()
df.groupby("EverBenched")["Salary"].mean().plot(kind="bar", color="red")
plt.title("Salary Comparison: EverBenched vs Not")
plt.xlabel("Ever Benched?")
plt.ylabel("Average Salary")
plt.show()

```

```python
# ---------- 18. City-wise Average Salary ----------
plt.figure()
df.groupby("City")["Salary"].mean().plot(kind="bar", color="cyan")
plt.title("Average Salary by City")
plt.xlabel("City")
plt.ylabel("Average Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------- 19. Gender Count ----------
plt.figure()
df["Gender"].value_counts().plot(kind="bar", color="magenta")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# ---------- 20. Department vs Age ----------
plt.figure()
df.groupby("Department")["Age"].mean().plot(kind="bar", color="brown")
plt.title("Average Age by Department")
plt.xlabel("Department")
plt.ylabel("Average Age")
plt.show()
```
