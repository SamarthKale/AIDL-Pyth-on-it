###1.4.1 Arithmetic Operators
They comprise:


*   Addition (+)
*   Subtraction (-)

*   Division (/)
*   Floor Division (//)
*   Modulo (%)

````python
x = 10
y = 3

# Add two operands
print('x + y =', x+y)

# Subtract right operand from the left
print('x - y =', x-y)

# Multiply two operands
print('x * y =', x*y)

# Divide left operand by the right one
print('x / y =', x/y)

# Floor division (quotient)
print('x // y =', x//y)

# Remainder of the division of left operand by the right
print('x % y =', x%y)

# Left operand raised to the power of right (x^y)
print('x ** y =', x**y)
````

### **Comparative Study: Conditional Operators (C vs Python)**

**What they do**  
Conditional operators compare two values and produce a boolean result (true/false). They control `if`/`else`, loops and expressions.

---

### **Operators (same symbols both languages)**
* `<`, `<=`  
* `>`, `>=`  
* `==`, `!=`

---

### **Quick comparison (concepts)**

*Boolean type*  
- C: `_Bool` or `int` (0 = false, non-zero = true).  
- Python: `bool` with values `True`/`False`.

*Equality vs identity*  
- C: `==` compares values. No separate identity operator.  
- Python: `==` compares values; `is` checks object identity (use `==` for value equality).

*Chained comparisons*  
- C: must write `a < b && b < c`.  
- Python: `a < b < c` is valid and reads naturally.

*Short-circuit evaluation*  
- Both languages short-circuit logical operators (`&&`/`||` in C, `and`/`or` in Python).

---

### **Examples**

**C**
```c
int a = 5, b = 7;
if (a < b) {
    printf("a is less than b\n");
}

if (a == 5) {
    printf("a equals 5\n");
}

/* chaining requires && */
if (a < b && b < 10) {
    printf("a < b < 10\n");
}


````python
a = 10
b = 20
if a < b:
    print("b is greater")
````

## 5. 🔁 The `while` Statement (Repetitive Execution)

The `while` loop is a control flow statement that repeats a block of code as long as a specified condition remains `True`.

### 1. Basic Structure Comparison

The core difference is Python's reliance on **indentation** and the **colon (`:`)** versus C's use of **curly braces (`{}`)** and **parentheses** around the condition.

#### **C Code Example**

```c
// C requires the condition in parentheses and an explicit update command inside the block.
int count = 0;
while (count < 5) {
    printf("C Count: %d\n", count);
    count++; // Explicit increment/update step is mandatory
}

````python
#Basic Format

#while (condition):
#    (task)
````

````python
#Example 1
number = 3
index = 1
limit = 10

while index <= limit:
    product = number*index
    print(number,'x', index,'=',product)
    index += 1
````

The `break` statement immediately **terminates the current loop** and resumes execution at the statement immediately following the loop body.

````python
for i in range(1, 11):
    if i == 5:
        break # Exit the loop entirely
    print(i, end=" ")
# Output: 1 2 3 4
# Execution continues here.
````

## 9. 📦 Modules in Python
  A **Module** in Python is simply a file ending with the extension `.py` that contains executable Python code (functions, classes, variables, runnable statements, etc.). They are the fundamental tool for code organization, promoting **reusability** and preventing namespace conflicts.
    
  The process of bringing code from one module into another is called **importing**.
    
  The `import` Statement
    
  To use the code from a module, you use the `import` keyword.
    
  python
  To import the entire 'math' module
  import math
  print(math.pi) # Access functions/variables using module.name
    
   To import specific functions/variables
  from math import sqrt, ceil
  print(sqrt(16)) # Can use the function directly


### Different Types of Modules

Python uses a well-defined hierarchy to manage code, classifying modules based on where they originate.

#### 1\. Built-in Modules (The Standard Library)

These modules are **pre-installed** with every standard Python installation. They provide essential functionalities for common tasks, meaning you don't need to download anything extra.

*   **Location:** Reside within the Python installation directory.
    
*   **Availability:** Always available on any Python environment (e.g., Colab, local machine).
    
*   **Examples:**
    
    *   **`math`**: Provides mathematical functions (trigonometry, exponentiation, constants like $\\pi$).
        
    *   **`random`**: Generates pseudo-random numbers.
        
    *   **`os`**: Provides interaction with the operating system (e.g., file paths, environment variables).
        
    *   **`sys`**: Provides access to system-specific parameters and functions.
        

#### 2\. External Modules (Third-Party Modules)

These modules are created by developers outside of the Python core team. They extend Python's capabilities for specialized purposes like web development, data science, or machine learning.

*   **Installation:** Must be explicitly downloaded and installed, typically using Python's package manager, **`pip`** (Python Installer Package).
    
    Bash
    
        # Example command to install an external module
        pip install numpy
    
*   **Availability:** Available only after installation.
    
*   **Examples:**
    
    *   **`numpy`**: For high-performance array operations (essential for data science).
        
    *   **`pandas`**: For data manipulation and analysis.
        
    *   **`requests`**: For making HTTP requests (e.g., fetching data from the web).
        

#### 3\. User-defined Modules

These are the modules you and your team write to organize your own project code. If you have a file named `calculator.py` with a function `add()`, then `calculator` is a user-defined module.

*   **Creation:** Simply write a Python file (`.py`).
    
*   **Usage:** You can import them directly into other files within the same project directory.
    
    *   If you have `main.py` and `calculator.py` in the same folder, `main.py` can use `import calculator`.
        
*   **Benefit:** Breaks large programs into small, manageable, logical components.

## 10. 🔌 External Modules (Third-Party Libraries)

External modules are collections of pre-written code created by the wider developer community. They are essential for extending Python's capabilities beyond its standard library, enabling specialized tasks like data analysis, web development, networking, and machine learning.

### 1. Key Characteristics

| Characteristic | Description |
| :--- | :--- |
| **Origin** | Created and maintained by third-party developers, not the Python core team. |
| **Specialization** | Highly specialized (e.g., handling complex numerical operations, building user interfaces). |
| **Requirement** | **Must be installed** separately, as they are not included in the basic Python installation. |

### 2. How to Install External Modules

External modules are managed using **pip** (Python's package installer), which connects to **PyPI** (The Python Package Index), the official repository for third-party Python software.

#### 📦 Using the `pip` Command

You execute `pip` commands from your system's command line or terminal (not from within the Python interpreter).

| Command | Purpose | Example |
| :--- | :--- | :--- |
| `pip install [package_name]` | Installs the latest stable version of a module. | `pip install requests` |
| `pip install [package_name]==[version]` | Installs a specific version of a module. | `pip install pandas==2.1.0` |
| `pip uninstall [package_name]` | Removes an installed module. | `pip uninstall numpy` |
| `pip list` | Shows all currently installed packages and their versions. | |
